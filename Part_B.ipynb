```python
# Copying data to another dataframe
data_copy = data_original.copy()
```

# Data Preprocessing

```python
# Removing duplicate records
data_copy.drop_duplicates(keep='first',inplace=True)
data_copy
# No duplicate values
```

```python
data_copy.info()
```

```python
print('Data columns with null values:\n',data_copy.isnull().sum())
```

```python
data_copy.drop(data_copy.iloc[:, 400:477], inplace=True, axis=1)
data_copy
```

```python
from statistics import mean
k = data_copy.columns.values.tolist()
for col in range(0,400):
    for row in range(0,1097):
        i = (data_copy.iloc[:,col].mean() + data_copy.iloc[row,:].mean())/2
        j = round(i,3)
        data_copy[k[col]].fillna(value=j,inplace=True, limit=1)
# Here we impute the missing ratings with a blend (50:50) of the arithmetic mean of each column and each row. 
```

```python
data_copy
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data_copy
```

```python
R_square = []
for i in range(0,399):
    X = data_copy.drop(k[i],axis=1)
    y = data_copy[k[i]]
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
    LR=LinearRegression()
    LR.fit(X_train,y_train)
    y_pred = LR.predict(X_test)
    R_square.append(r2_score(y_test, y_pred)) 
```

```python
R_square
```

```python
for j in range(0,399):
    X = data_copy.drop(k[j],axis=1)
    y = data_copy[k[j]]
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
    LR=LinearRegression()
    LR.fit(X_train,y_train)
    y_pred = LR.predict(X_test)
    if max(R_square) == r2_score(y_test, y_pred):
        print("Maximum R2 value is",r2_score(y_test, y_pred))
        print("The movie that predicts ratings the best is",k[j])
```

```python
print("The average COD (Coefficient of determination) of those 400 simple linear regression models is",mean(R_square))
```

```python
# Distribution of COD or R2 score
plt.figure(figsize=(10,5))
sns.histplot(R_square, kde=True)
plt.title("Distribution of 'COD or R2 score'")
plt.xlabel("R2_score")
plt.ylabel("Frequency")
plt.show()
```

```python
best_pred_movies = []
for i in range(-10,0):
    for j in range(0,399):
        X = data_copy.drop(k[j],axis=1)
        y = data_copy[k[j]]
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
        LR=LinearRegression()
        LR.fit(X_train,y_train)
        y_pred = LR.predict(X_test)
        if R_square[i] == r2_score(y_test, y_pred):
            best_pred_movies.append(k[j])
```

```python
best_pred_movies
```

```python
worst_pred_movies = []
for i in range(0,10):
    for j in range(0,399):
        X = data_copy.drop(k[j],axis=1)
        y = data_copy[k[j]]
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
        LR=LinearRegression()
        LR.fit(X_train,y_train)
        y_pred = LR.predict(X_test)
        if R_square[i] == r2_score(y_test, y_pred):
            worst_pred_movies.append(k[j])
```
